```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling.head(10)
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

```python
error
```

```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
```

```python
combined.head(10)
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"}
mapping = MissingDict(**map_values)
```

```python
combined["new_team"] = combined["team"].map(mapping)
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
```

```python
merged.shape
merged.head(30)
```

```python
merged[(merged["predicted_x"] == 1) & (merged["predicted_y"] ==0)]["actual_x"].value_counts()
```

```python
# Count wins for each team in 'team_x'
team_x_wins = merged[merged['result_x'] == 'W']['team_x'].value_counts()

# Count wins for each team in 'team_y'
team_y_wins = merged[merged['result_y'] == 'W']['team_y'].value_counts()

# Identify the team with the most wins in 'team_x'
most_wins_team_x = team_x_wins.idxmax()
most_wins_count_x = team_x_wins.max()

# Identify the team with the most wins in 'team_y'
most_wins_team_y = team_y_wins.idxmax()
most_wins_count_y = team_y_wins.max()

print(f"The team with the most wins as 'team_x' is {most_wins_team_x} with {most_wins_count_x} wins.")
print(f"The team with the most wins as 'team_y' is {most_wins_team_y} with {most_wins_count_y} wins.")
```

```python
results = pd.DataFrame(columns=['Team', 'W', 'D', 'L', 'Points'])

# Calculate results for team_x
team_x_results = merged.groupby('team_x')['result_x'].value_counts().unstack().fillna(0)
team_x_results.columns = [col if col in ['W', 'D', 'L'] else 'Other' for col in team_x_results.columns]
team_x_results = team_x_results[['W', 'D', 'L']]
team_x_results = team_x_results.reset_index().rename(columns={'team_x': 'Team'})

# Calculate results for team_y
team_y_results = merged.groupby('team_y')['result_y'].value_counts().unstack().fillna(0)
team_y_results.columns = [col if col in ['W', 'D', 'L'] else 'Other' for col in team_y_results.columns]
team_y_results = team_y_results[['W', 'D', 'L']]
team_y_results = team_y_results.reset_index().rename(columns={'team_y': 'Team'})

# Combine results
combined_results = pd.concat([team_x_results, team_y_results]).groupby('Team').sum().reset_index()

# Calculate points
combined_results['Points'] = combined_results['W'] * 3 + combined_results['D'] * 1

# Convert columns to integer type
combined_results = combined_results.astype({'W': 'int', 'D': 'int', 'L': 'int', 'Points': 'int'})

# Convert DataFrame to list of dictionaries
results_list = combined_results.to_dict('records')

# Sort the list by 'Points' in descending order
sorted_results_list = sorted(results_list, key=lambda x: x['Points'], reverse=True)

# Convert the sorted list back to DataFrame
sorted_combined_results = pd.DataFrame(sorted_results_list)

# Print the sorted DataFrame without the index
print(sorted_combined_results.to_string(index=False))
```
